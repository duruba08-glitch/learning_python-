```python
def sample(tup):
       maxi=max(tup)
       mini=min(tup) 
       sum=0
       for i in tup:
                sum=sum+i
       print(maxi)
       print(mini)
       print(sum)

         
    
tups=(1,2,3,4,5,6)
sample(tups)
```
