```python
##1
def fib(n,memo={}):
    if n in memo:
        return memo[n]
    if n==0:
        return 0
    if n==1:
        return 1
    memo[n]=fib(n-1,memo) + fib(n-2,memo)
    return memo[n]

print(fib(10))
print(fib(9))
print(fib(11))
print(fib(4))
print(fib(2))
print(fib(3))
print(fib(6))
print(fib(1))


```

```python
#6

import time
def deco(func):
    def wrap(a,b):
        start=time.time()
        res=func(a,b)
        end=time.time()
        time_taken=end - start
        print(res)
        print("time taken:",time_taken)
         
    return wrap


@deco
def add(a,b):
    return a+b

add(2,1)
```

```python
def feb_gene(x):
    a=0
    b=1
    num=0
    count=0
    while count<x:
        yield(a)
        num=a+b
        a,b=b,num
        count+=1
    
        

for i in feb_gene(10):
    print(i)
```
