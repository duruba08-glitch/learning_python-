```python
##1
def fib(n,memo={}):
    if n in memo:
        return memo[n]
    if n==0:
        return 0
    if n==1:
        return 1
    memo[n]=fib(n-1,memo) + fib(n-2,memo)
    return memo[n]

print(fib(10))
print(fib(9))
print(fib(11))
print(fib(4))
print(fib(2))
print(fib(3))
print(fib(6))
print(fib(1))


```
