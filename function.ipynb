```python
#temperature conversion
def temp_conversion(temp,unit):
    if unit=="F":
        res=5/9 *(temp - 32)
    else:
        res=9/5*temp+32
    
    return res

temp_conversion(23,"F")
temp_conversion(23,"C")
        
    
```

```python
#checking a password is strong or not
def isstrong_pass(password):
    if len(password)<=8:
        return False
    elif not any(char.isdigit() for char in password):
        return False
    elif not any(char.isupper() for char in password):
        return False
    elif not any(char.islower() for char in password):
        return False 
    elif not any(char in "! @ # $ % ^ & *" for char in password):
        return False
    else:
        return True
    
isstrong_pass("123B@ll")

```

```python
#Total cost of items in a shopping casrt
def total_cost(shoping_cart):
    total=0
    for  item in shoping_cart:
            total=total+item['price']*item['quantity']
    return total
                  

total_cost(
[{'name': 'Apple', 'price': 0.5, 'quantity' :4},
{'name': 'Banana', 'price':0.3, 'quantity' :6}, 
{'name' : 'Orange', 'price': 0.7, 'quantity':3}]
)
```

```python
#check if a string is palindrome or not

def is_palin(S):
    S=S.lower().replace(" ","")
    return S==S[::-1]

      

is_palin("A man a plan a canal panama")
```

```python
#factorial of a numbe using recursion
def fact(n):
    if n==0:
        return 1
    else:
        return n * fact(n-1)

fact(3)
```
