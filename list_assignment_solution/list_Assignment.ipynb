```python
def flaten(list):
    new_list=[]
    for i in list:
        for j in i:
            new_list.append(j)
    return new_list


List=[[1,2,3],
      [4,5,6],
      [3,5,9]
      ]


for row in List:
    print(row)
flaten(List)
```

```python
def rotate(Lst,k,direction):
    k=k%len(Lst)
    
    if direction=="left":
        a=Lst[k:]+ Lst[:k]
        return a
    elif direction=="right":
        b=Lst[-k:]+Lst[:-k]
        return b 
ls=[1,2,3,4,5,]
print(rotate(ls,2,"left"))
print(rotate(ls,2,"right"))
```

```python
def intersect(lst1,lst2):
    new_list=[]
    for i in lst1:
        for j in lst2:
            if i==j:
                new_list.append(i)

    return set(new_list)

a=[1,2,3,4,5,7]
b=[3,2,7,4,5,7]
intersect(a,b)

        
```
